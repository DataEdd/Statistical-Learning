```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the advertising dataset

auto_data = pd.read_csv('Advertising.csv')

# Display the first few rows of the dataset
print(auto_data.head())

# Summary statistics
print(auto_data.describe())
```

```python
auto_data['Budget'] = auto_data['TV'] + auto_data['radio'] + auto_data['newspaper']
```

```python
sns.pairplot(auto_data)
```

```python
sns.relplot(x='Budget', y='sales', data=auto_data)
plt.show()
```

```python
from matplotlib.pyplot import errorbar


sns.relplot(x='Budget', y='sales', data=auto_data, kind='line', errorbar = 'sd')
plt.show()
```

```python
sns.relplot(data=auto_data, x = 'TV',y = 'radio', size = 'sales', hue = 'sales')
```

```python
sns.relplot(data=auto_data, x = 'newspaper',y = 'radio', size = 'sales', hue = 'sales')
```
